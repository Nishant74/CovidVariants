## Section : 01
### Total Number of sequences on the basis of month

```python
gp=df.groupby("month")["num_sequences_total"].sum()
```

```python
t=gp.reset_index()
```

```python
t
```

```python
t1=t.sort_values(by=["num_sequences_total"],ascending=False)
```

```python
a=t.month
b=t.num_sequences_total
```

```python
plt.figure(figsize=(56,24))
plt.title(" Total Weekly Number of Sequences Carried Out By Month",color="gold",fontsize=78)
plt.xlabel("Month Number",color="gold",fontsize=36)
plt.ylabel(" Number of Sequences in Million",color="gold",fontsize=36)
plt.xticks(fontsize=48,color="lightsalmon")
plt.yticks(fontsize=48,color="lightsalmon")
sns.lineplot(a,b,linewidth=20)
plt.show()
```

## Section  Conclusion :

## Section :04

```python
plt.figure(figsize=(24,12))
plt.title('Top 3 Variants By Month')
plt.title(" Total Weekly Number of Sequences Carried Out By Top 3 Variants",color="gold",fontsize=36)
plt.xticks(fontsize=24,color="w")
plt.yticks(fontsize=24,color="w")
df.loc[df['variant'] == 'non_who'].groupby('month')['num_sequences_total'].agg('sum').plot(label='Non Who')
df.loc[df['variant'] == 'others'].groupby('month')['num_sequences_total'].agg('sum').plot(label='Others')
df.loc[df['variant'] == 'Delta'].groupby('month')['num_sequences_total'].agg('sum').plot(label='Delta')
plt.ylabel("Total Number of Sequences in Million",color="gold",fontsize=24)
plt.xlabel("Month",color="gold",fontsize=24)
plt.legend(fontsize=36)
plt.show()
```
